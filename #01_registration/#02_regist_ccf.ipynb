```python
# import libraries

import os
from os.path import join as pjoin
from tifffile import *
from wfield import *
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set save folder

localdisk = "base_path"
dat_path = "img_path"
ccf_path = "ccf_path"
template_path =  f"{dat_path}/template_dir"
print(dat_path)
print(localdisk)
fs = 20 # per channel
time = 40
```

```python
# rewrite microscope and camera condition you recorded

import pandas as pd

binning = 4
x = 2*0.7                     # objective lens & imaging lens
resolution = 0.0065*binning*x # mm/pixel
```

```python
# import average image

# The image variable is supposed to be single frame with shape (H, W)
image = np.load(glob(pjoin(template_path, "*frames_average.npy"))[0])[0]
H, W = image.shape
bregma_offset = (np.array(image.shape[::-1])/2).astype('int')
print(f"bregma_offset:{bregma_offset}")
print(f"image.shape:{image.shape}")
plt.imshow(image)
```

```python
# import allen brain map

## if the refrence folder is none or not specified,
## then folders are created to ~/wfield/ and related files are dowloaded from the j.couto wfield and saved there.
## Check wfield/allen.py if you are unclear about the 'allen_load_reference' function
ccf_regions, proj, brain_outline = allen_load_reference(annotation_dir=ccf_path,
                                                        reference_name="dorsal_cortex")

# Get or load landmarks, these are for the allen dorsal cortex dataset
landmarks = {'x': [-1.95, 0, 1.95, 0],
             'y': [-3.45, -3.45, -3.45, 3.2],
             'name': ['OB_left', 'OB_center', 'OB_right', 'RSP_base'],
             'color': ['#fc9d03', '#0367fc', '#fc9d03', '#fc4103']}
landmarks = pd.DataFrame(landmarks)

# Load holoviews with the bokeh backend
import holoviews as hv
hv.extension('bokeh')
```

```python
# check landmarks on the brain map

landmarks = {'x': [-1.95, 0, 1.95, 0],
             'y': [-3.45, -3.45, -3.45, 3.2],
             'name': ['OB_left', 'OB_center', 'OB_right', 'RSP_base'],
             'color': ['#fc9d03', '#0367fc', '#fc9d03', '#fc4103']}

wid, lmark_wid = hv_adjust_reference_landmarks(landmarks,ccf_regions)
wid
```

```python
# check location of the landmarks

landmarks = pd.DataFrame(lmark_wid.data)[['x','y','name','color']]
save_allen_landmarks(landmarks,
                     pjoin(template_path, "analysis_landmarks.json"),
                     resolution = resolution,
                     bregma_offset = bregma_offset)
landmarks
```

```python
# check resolution and bregma

lmarks = load_allen_landmarks(pjoin(template_path, "analysis_landmarks.json"))
landmarks = lmarks['landmarks']
if 'landmarks_match' in lmarks.keys():
    landmarks_match = lmarks['landmarks_match']
else:
    landmarks_match = None
bregma_offset = lmarks['bregma_offset']
resolution = lmarks['resolution']
print(f"resolution:{resolution}")
print(f"bregma_offset:{bregma_offset}")
```

```python
# place 4 landmarks on your image

wid,lmark_wid,landmarks_im = hv_adjust_image_landmarks(image,landmarks,
                                                       landmarks_match = landmarks_match,
                                                       bregma_offset = bregma_offset,
                                                       resolution = resolution)
print(f"place the landmark manually")
wid

## After running this cell, it will be essential to mannually adjust the landmarks in the figure to the right position.
## You can do this by clicking and dragging the landmarks.
```

```python
# Get the similarity transform and plot the result

landmarks_match = pd.DataFrame(lmark_wid.data)
M = allen_transform_from_landmarks(landmarks_im,landmarks_match)
hv_show_transformed_overlay(image, M, ccf_regions,bregma_offset=bregma_offset,resolution = resolution)
```

```python
# check how much it has been changed from raw data

from skimage.transform import SimilarityTransform
# inverse_M = SimilarityTransform(matrix=iM)
save_allen_landmarks(landmarks,
                     filename = pjoin(template_path, "analysis_landmarks.json"),
                     landmarks_match=landmarks_match,
                     transform=M,
                     resolution = resolution,
                     bregma_offset = bregma_offset)

print('''Transform:
    - scale {0}
    - translation {1}
    - rotation {2}'''.format(M.scale,M.translation,np.rad2deg(M.rotation)))
```
