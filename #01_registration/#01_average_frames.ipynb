```python
# import libraries

import os
from os.path import join as pjoin
from tifffile import *
from wfield import *
from glob import glob
from tqdm import tqdm
```

```python
# set save folder

localdisk = "base_path"
dat_path = "img_path"
ccf_path = "ccf_path"
template_path =  f"{dat_path}/template_dir"
print(dat_path)
print(localdisk)
fs = 20 # per channel
time = 40
```

```python
# import your folders for every cycle

datfolders = [pjoin(dat_path,i) for i in os.listdir(dat_path) if os.path.isdir(pjoin(dat_path,i))]
datfolders = sorted(datfolders)
for i in datfolders:
    print(i)
```

```python
# save average images for each excitation wave length

import numpy as np

for folder in tqdm(datfolders):
    files = sorted(glob(pjoin(folder, "*.tif")))
    for file in files:
        dat = memmap(file)
        if len(dat.shape) == 4:
            frames_average = np.mean(dat, axis=0)
        elif len(dat.shape)==3:
            nframe, H, W = dat.shape
            nframe = int(nframe/2)
            dat = dat.reshape((nframe, 2, H, W))
            frames_average = np.mean(dat, axis=0)
        else:
            raise ValueError(f"Something is wrong with your tiff data shape.\nYour data shape is {dat.shape}")
        save_path = file.rstrip(".tif") + "_frames_average.npy"
        np.save(pjoin(folder, save_path), frames_average)
```
