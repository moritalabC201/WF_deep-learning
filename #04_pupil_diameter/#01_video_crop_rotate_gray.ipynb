```python
import os
import cv2
from moviepy.editor import VideoFileClip, vfx

def select_roi(frame):
    roi = cv2.selectROI("Select ROI", frame, False)
    cv2.destroyAllWindows()
    return roi

folder_path = "mp4_path"

mp4_files = [f for f in os.listdir(folder_path) if f.endswith(".mp4")]
if not mp4_files:
    print("mp4 is not found")
    exit()

first_mp4 = os.path.join(folder_path, mp4_files[0])

video = cv2.VideoCapture(first_mp4)
ret, frame = video.read()
video.release()


print("select region")
crop_area = select_roi(frame)

print(f"selected region: x={crop_area[0]}, y={crop_area[1]}, width={crop_area[2]}, height={crop_area[3]}")

for filename in mp4_files:
    input_path = os.path.join(folder_path, filename)
    output_path = os.path.join(folder_path, f"processed_{filename}")
    
    video = VideoFileClip(input_path)
    cropped_video = video.crop(x1=crop_area[0], y1=crop_area[1], 
                               width=crop_area[2], height=crop_area[3])
    
    rotated_video = cropped_video.rotate(90)

    grayscale_video = rotated_video.fx(vfx.blackwhite)

    grayscale_video.write_videofile(output_path, fps=20)

    video.close()
    cropped_video.close()
    rotated_video.close()
    grayscale_video.close()

print("completed")
```
